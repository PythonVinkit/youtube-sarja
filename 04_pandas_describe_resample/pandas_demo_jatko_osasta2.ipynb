## Importit

```python
from datetime import datetime
import pandas as pd
```

## Luetaan osaketiedot käyttäen hyödyksi read_csv-funktion date_parser-parametria.
date_parser-parametrille on annettava funktio, joka käsittelee yhden aikaleiman (aikasarakkeen elementin) kerrallaan.

Videolla Junnu kysyi, yritetäänkö jokaista DataFramen elementtiä syöttää vuorollaan argumenttina annetulle funktiolla. Vastaus on, että Pandas yrittää suorittaa funktion kolmella tapaa aloittaen ensimmäisestä ja siirtyen eteenpäin, mikäli suoritus aiheuttaa poikkeuksen (Exception)
 1. Anna funktiolle argumentiksi parse_dates-parametrissa määritellyt taulukot (Arrays)
 2. Liitä rivittäin yhteen määritellyt sarakkeet ja anna yhdistetty taulukko argumenttina
 3. Kutsu funktiota erikseen jokaiselle DataFramen riville yksitellen ja käytä argumenttina date_parser-parametrin sarakkeessa olevaa merkkijonoa (esim. parse_dates = ['PVM'])

 Alla olevassa koodissa siis ensimmäinen ja toinen kohta tuottavat poikkeuksen ja kolmas halutun toteutuksen. Pienellä testaamisella selvisi, että myös index_col ajetaan date_parserin läpi, vaikka en löytänyt dokumentaatiossa mainintaa tästä.

Englanniksi sama selitys löytyy [tästä](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html) linkistä (kohta date_parser)

```python
date_parser = lambda x: datetime.strptime(x, '%d/%m/%Y')
data = pd.DataFrame()
data = pd.read_csv('../data/osakkeet.csv', date_parser=date_parser, index_col='PVM')
```

## Käännetään data-DataFramen rivit ylösalaisin

```python
data = data.iloc[::-1]
data.head()
```

## Testataan loc- ja iloc-funktiota
